# Find month over month growth comparison for 2022 and 2023 sales eg : jan 2022 vs jan 2023

```python
df1.head(2)
```

```python
a= df1.groupby(['Month', 'Year']).sum(['Total_sales'])['Total_sales']
df2= pd.DataFrame(a)
```

```python
pivot_table = df2.pivot_table(index = 'Month', columns = 'Year', values = 'Total_sales', aggfunc= 'sum')
pivot_table
```

```python
pivot_table = pivot_table.reset_index()
pivot_table
```

```python
pivot_table['yoy_growth'] = pivot_table[2022] - pivot_table[2023]
pivot_table['%_growth'] = ((pivot_table[2022] - pivot_table[2023])/pivot_table[2022])*100
```
